```python
import numpy as np
import os 
import h5py
import scipy.io
```

```python
data_dir = '/data/cees/gegewen/'
```

```python
# load data S
S_data = scipy.io.loadmat('/data/cees/gegewen/cs231n_project_data/sat_map.mat')['sat_map']
print('S_data shape', S_data.shape)
```

```python
# load data inj
inj_data = scipy.io.loadmat('/data/cees/gegewen/cs231n_project_data/inj_map.mat')['inj_map']
print('inj_data shape', inj_data.shape)
```

```python
# load data P
P_data = scipy.io.loadmat('/data/cees/gegewen/cs231n_project_data/pressure_map.mat')['pressure_map']
print('Pressure_data shape', P_data.shape)
```

```python
# reshape data S
S = S_data
print('S shape', S.shape)
S = S.reshape(128, 128, 4, 500,1)
print('S shape', S.shape)
```

```python
# reshape data inj
inj = inj_data
print('inj shape', inj.shape)
inj = np.repeat(inj[:, :, np.newaxis, :], 4, axis = 2)
print('inj shape', inj.shape)
inj = inj.reshape(128, 128, 4, 500, 1)
print('inj shape', inj.shape)
inj = inj.swapaxes(0,1)
```

```python
inj_day = np.zeros_like(inj)
inj_day[:, :, 0, :, :] = inj[:, :, 0, :, :].astype('float64') * 25
inj_day[:, :, 1, :, :] = inj[:, :, 1, :, :].astype('float64') * 50
inj_day[:, :, 2, :, :] = inj[:, :, 2, :, :].astype('float64') * 75
inj_day[:, :, 3, :, :] = inj[:, :, 3, :, :].astype('float64') * 100
```

```python
# reshape data P
P = P_data
print('P shape', P.shape)
P = P.reshape(128, 128, 4, 500, 1)
print('P shape', P.shape)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt 

day_idx = 3

for k_idx in range (0,5):
    plt.set_cmap('jet')

    plt.figure(figsize=(12, 3))
    plt.subplot(1,3,1)
    plt.imshow(S[:, :, day_idx, k_idx,0])
    plt.title('SG')
    plt.plot()
    plt.colorbar(fraction=0.046)
    
    plt.subplot(1,3,2)
    plt.imshow(inj_day[:, :, day_idx, k_idx, 0])
    plt.title('inj (day)')
#     plt.clim([-0.2, 0.2])
    plt.plot()
    plt.colorbar(fraction=0.046)

    plt.subplot(1,3,3)
    plt.imshow(P[:, :, day_idx, k_idx, 0])
#     plt.clim([0, 200])
    plt.title('P')
    plt.plot()
    plt.colorbar(fraction=0.046)

    inj[k_idx, :, 1, 0] = 0
    
    plt.tight_layout()
    plt.show()
```

```python
inj_S_P = np.concatenate([inj_day, S, P], axis = 4)
print('package shape is ', inj_S_P.shape)
inj_S_P = inj_S_P.swapaxes(4,3)
print('package shape is ', inj_S_P.shape)
inj_S_P = inj_S_P.reshape(128, 128, 4, 3, 500, 1)
```

```python
inj_S_P_data = np.concatenate([inj_S_P[..., :250, :], inj_S_P[..., 250:, :]], axis = 5)
print('package shape is ', inj_S_P_data.shape)
inj_S_P_data = inj_S_P_data.swapaxes(4,5)
print('package shape is ', inj_S_P_data.shape)
inj_S_P_data = inj_S_P_data.swapaxes(2,3)
print('package shape is ', inj_S_P_data.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

day_idx = 3
t_idx = 0

for k_idx in range (0,5):

    plt.set_cmap('jet')

    plt.figure(figsize=(12, 3))
    plt.subplot(1,3,1)
    plt.imshow(inj_S_P_data[:, :,0, day_idx, t_idx, k_idx])
    plt.plot()
    plt.colorbar(fraction=0.046)
    
    plt.subplot(1,3,2)
    plt.imshow(inj_S_P_data[:, :,1, day_idx, t_idx, k_idx])
    plt.plot()
    plt.colorbar(fraction=0.046)

    plt.subplot(1,3,3)
    plt.imshow(inj_S_P_data[:, :,2, day_idx, t_idx, k_idx])
    plt.plot()
    plt.colorbar(fraction=0.046)

    plt.tight_layout()
    plt.show()
```

```python
hw = h5py.File('/data/cees/gegewen/cs231n_project_data/inj_S_P_data.hdf5', 'w')
hw.create_dataset('inj_S_P_data', data = inj_S_P_data)
hw.close()
```
