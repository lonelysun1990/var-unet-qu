```python
input_shape = (128, 128, 2)
output_shape = (128, 128, 1)
```

```python
## Original v-unet implementation
# enc_up, hidden_shapes = vunet.create_enc_up(input_shape, n_scales = 5, n_res_blk = 1)
# # enc_up.summary()

# latent_shape = hidden_shapes[-1]

# enc_down = vunet.create_enc_down(latent_shape)
# # enc_down.summary()

# dec_up = vunet.create_dec_up(output_shape, n_scales = 5, n_res_blk = 1) # only c input
# # dec_up.summary()

# dec_down = vunet.create_dec_down(hidden_shapes, latent_shape, n_scales = 5, n_res_blk = 1)
# # dec_down.summary()
```

```python
# change the role of enc_up and dec_up
enc_up, hidden_shapes = vunet.create_enc_up(output_shape, n_scales = 5)
enc_down = vunet.create_enc_down(hidden_shapes[-1])
dec_up = vunet.create_dec_up(input_shape, n_scales = 5) # only c input
dec_down = vunet.create_dec_down(hidden_shapes, hidden_shapes[-1], n_scales = 5)
```

#### Train forward pass v-unet

```python
# xc = Input(shape=input_shape) # perm + well location
# c = Input(shape=output_shape) # well location

# hs = enc_up(xc)
# z_posterior_sample, z_posterior_mean = enc_down(hs[-1])
# gs = dec_up(c)
# # x_hat, z_piror_sample, z_piror_mean = dec_down([z_posterior_sample, *gs])
# x_hat, z_piror_sample, z_piror_mean = dec_down([z_posterior_mean, *gs])
```

```python
xc = Input(shape=input_shape) # perm + well location
c = Input(shape=output_shape) # well location

hs = enc_up(c)
z_posterior_sample, z_posterior_mean = enc_down(hs[-1])
gs = dec_up(xc)
x_hat, z_piror_sample, z_piror_mean = dec_down([z_posterior_mean, *gs])
```

```python
def likelihood_loss(x, x_hat):
    '''
    Reconstruction loss, L2 norm
    
    '''
    v = 0.1
    return K.mean(K.sum((K.batch_flatten(x) - K.batch_flatten(x_hat)) ** 2 / (2*v), axis=-1))

def latent_kl_loss(q, p):
    mean1 = q
    mean2 = p
    
    kl = 0.5 * (mean2 - mean1)**2
    kl = K.sum(kl, axis = [1,2,3])
    kl = K.mean(kl)
    return kl
```

```python
x = Input(shape=output_shape) # target saturation

likelihood_loss_value = likelihood_loss(x, x_hat)
kl_loss = latent_kl_loss(z_posterior_mean, z_piror_mean)
kl_weight = 5.0
total_loss = likelihood_loss_value + kl_weight * kl_loss
```

```python
epoch = 300
learning_rate = 5e-4
train_nr = train_x.shape[0]
test_nr = test_x.shape[0]
batch_size = 2
num_batch = int(train_nr/batch_size) 
```

```python
opt = Adam(lr=learning_rate)

trainable_weights = enc_up.trainable_weights + enc_down.trainable_weights + dec_up.trainable_weights + dec_down.trainable_weights

updates = opt.get_updates(total_loss, trainable_weights)

iterate = K.function([xc, c, x], [total_loss, likelihood_loss_value, kl_loss], updates=updates)

eval_loss = K.function([xc, c, x], [total_loss])
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)

train_shuffled_idx = np.random.permutation(train_x.shape[0])
test_shuffled_idx = np.random.permutation(test_x.shape[0])

# print(train_shuffled_idx)
# print(test_shuffled_idx)
train_x_shuffled = train_x[train_shuffled_idx,:,:,:]
train_y_shuffled = train_y[train_shuffled_idx,:,:,:]

test_x_shuffled = test_x[test_shuffled_idx,:,:,:]
test_y_shuffled = test_y[test_shuffled_idx,:,:,:]
# print(train_x_shuffled.shape)

train_x_y_hat0 = np.concatenate((train_y, train_x[:,:,:,[1]]), axis=-1)
# train_x00 = train_x_y_hat0[:,:,:,[0]]
# train_y_hat00 = train_x_y_hat0[:,:,:,[1]]
train_y_hat0 = train_x_shuffled[:,:,:,[1]]
train_x_hat0 = train_y_shuffled


test_x_y_hat0 = np.concatenate((test_y, test_x[:,:,:,[1]]), axis=-1)
# test_x00 = test_x_y_hat0[:,:,:,[0]]
# test_y_hat00 = test_x_y_hat0[:,:,:,[1]]
test_y_hat0 = test_x_shuffled[:,:,:,[1]]
test_x_hat0 = test_y_shuffled

# print(train_x00.shape)
# print(train_y_hat00.shape)
print(train_y_hat0.shape)
print(train_x_hat0.shape)

# print(test_x00.shape)
# print(test_y_hat00.shape)
print(test_y_hat0.shape)
print(test_x_hat0.shape)
```

```python
for k in range(9,12):
    plt.figure(k,figsize=(15,3))
    
    plt.subplot(1,5,1)
    plt.title('x (train)')
    plt.imshow(train_x_y_hat0[k, :, :, 0])
    plt.colorbar(fraction=0.046)
    
    plt.subplot(1,5,2)
    plt.title('y_hat (train)')
    plt.imshow(train_x_y_hat0[k, :, :, 1] * 200)
    plt.clim(0, 200)
    plt.colorbar(fraction=0.046)
    
    plt.subplot(1,5,3)
    plt.title('x (true)')
    plt.imshow(train_x_hat0[k, :, :, 0].reshape(128, 128))
#     plt.clim([0, 0.6])
    plt.colorbar(fraction=0.046)
    
    plt.subplot(1,5,4)
    plt.title('y_hat (true)')
    plt.imshow(train_y_hat0[k, :, :, 0] * 200)
    plt.clim(0, 200)
    plt.colorbar(fraction=0.046)
    
    plt.subplot(1,5,5)
    rel_err = np.abs((train_x_hat0[k, :, :, 0] - train_x_y_hat0[k, :, :, 0]))
    plt.imshow(rel_err) 
    plt.title('Abs Error')
    plt.colorbar(fraction=0.046)
    plt.tight_layout()
    plt.show()
```

```python
## log for tensorboard
def write_summary(value, tag, summary_writer, global_step):
    """Write a single summary value to tensorboard"""
    summary = tf.Summary()
    summary.value.add(tag=tag, simple_value=value)
    summary_writer.add_summary(summary, global_step)

    
summary_writer = tf.summary.FileWriter('logs/' + current_time)
```

```python
for e in range(epoch):
    for ib in range(num_batch):
        ind0 = ib * batch_size

        x_y_hat0_batch = train_x_y_hat0[ind0:ind0+batch_size, ...]
        y_hat0_batch = train_y_hat0[ind0:ind0+batch_size, ...]
        x_hat0_batch = train_x_hat0[ind0:ind0+batch_size, ...]
        
        train_losses = iterate([x_y_hat0_batch, y_hat0_batch, x_hat0_batch])
        ## -- Equivalent to:
        # train_losses = tf.session.run(feed_dict={xc: x_y_hat0_batch, c: y_hat0_batch, x: x_hat0_batch}, ...
        #                               fetches= [total_loss, likelihood_loss_value, kl_loss])

        if ib % 1 == 0:
            print('Epoch %d/%d, Batch %d/%d, Total Loss %f, Likelihood Loss %f, KL Loss %f' % (e+1, epoch, ib+1, num_batch, train_losses[0], train_losses[1], train_losses[2]))
        
        eval_losses = eval_loss([test_x_y_hat0, test_y_hat0, test_x_hat0])
        
        write_summary(eval_losses[0], 'eval_loss', summary_writer, e) # log for tensorboard
        write_summary(train_losses[0], 'train_loss', summary_writer, e) # log for tensorboard

    print('====================================================')
    print('\n')
    print('Epoch %d/%d, Train Total Loss %f, Eval Total Loss %f' % (e + 1, epoch, train_losses[0], eval_losses[0]))
    print('\n')
    print('====================================================')

    
enc_up.save_weights(output_dir + 'enc_up_' + case_string + '_%dtrain_lr%.0e_ep%d.h5' % (train_nr, learning_rate, epoch))
enc_down.save_weights(output_dir + 'enc_down_' + case_string + '_%dtrain_lr%.0e_ep%d.h5' % (train_nr, learning_rate, epoch))
dec_up.save_weights(output_dir + 'dec_up_' + case_string + '_%dtrain_lr%.0e_ep%d.h5' % (train_nr, learning_rate, epoch))
dec_down.save_weights(output_dir + 'dec_down_' + case_string + '_%dtrain_lr%.0e_ep%d.h5' % (train_nr, learning_rate, epoch))
```
